```python
from faker import Faker
import pandas as pd
fake=Faker()
```

```python
Faker.seed(5)
n=50
patients = pd.DataFrame(
{
    "SSN" : [fake.ssn() for i in range(n)],
    "Name" : [fake.name() for i in range(n)],
    "Street" : [fake.street_address() for i in range(n)],
    "City" : [fake.city() for i in range(n)],
    "State" : [fake.state_abbr() for i in range(n)],
    "Zip" : [fake.zipcode_in_state() for i in range(n)],
    "Age" : [fake.random_int(18,80) for i in range(n)],
    "Cell" : [fake.phone_number() for i in range(n)]
})

patients
```

```python
Faker.seed(10)
n=15
departments = ["Orthopaedics", "Neurology", "Cardiology", "Psychiatry", "Skin", "Plastic Surgery", "Surgery", "Gynaecology", "Obstetrics", "Paediatrics", "Infectious Disease", "Burn Center", "Gastroenterology", "Haematology", "Renal"]
doctors = pd.DataFrame(
{
    "SSN" : [fake.ssn() for i in range(n)],
    "Name" : [fake.name() for i in range(n)],
    "Department":[departments[i] for i in range(len(departments))]
})

doctors
```

```python
def ncycles(iterable, n):
    "Returns the sequence elements n times"
    from itertools import chain, repeat
    return chain.from_iterable(repeat(iterable, n)) 
    # the general recipe wraps iterable in tuple()
    # to ensure you can walk it multiple times
    # here we know it is always a list
```

```python
Faker.seed(15)
n=30
departments = ["Orthopaedics", "Neurology", "Cardiology", "Psychiatry", "Skin", "Plastic Surgery", "Surgery", "Gynaecology", "Obstetrics", "Paediatrics", "Infectious Disease", "Burn Center", "Gastroenterology", "Haematology", "Renal"]
nurses = pd.DataFrame(
{
    "SSN" : [fake.ssn() for i in range(n)],
    "Name" : [fake.name() for i in range(n)],
    "Department":[i for i in ncycles(departments, 2)]
})

nurses
```
